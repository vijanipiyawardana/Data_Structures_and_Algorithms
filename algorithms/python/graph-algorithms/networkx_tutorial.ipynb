```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# simple undirected graph
G = nx.Graph() 
```

```python
# simple directed graph
G = nx.DiGraph()
```

```python
# undirected graph, multiple edges between nodes
G = nx.MultiGraph()
```

```python
# directed graph, multiple edges between nodes
G = nx.MultiDiGraph()
```

```python
# adding edges between nodes
G = nx.Graph()
G.add_edge(1, 2)
```

```python
# edges with weight: can be speed, cost, etc. higher the weight more difficult to get from node 2 to 3
G.add_edge(2, 3, weight = 0.9)
```

```python
# nodes with 'letter' names
G.add_edge("A", "B")
```

```python
# can have edge to same node from it
G.add_edge("B", "B")
```

```python
# can have individual nodes
G.add_node("C")
```

```python
# anything pass as parameter is an object
# can pass print function, but note calling it here
G.add_node(print)
```

```python
# visualize the graph
# show with names/labels of the node

nx.draw_spring(G, with_labels = True)
plt.show()
```

```python
# create a graph from an edge list. edge list is given as a tuple.
# edge list representation of a graph

edge_list = [(1, 2), (2, 3), (3, 4), (3, 5), (4, 6), (6, 7)]
G = nx.from_edgelist(edge_list)

# OR
# G = nx.Graph()
# G.add_edges_from(edge_list)

nx.draw_spring(G, with_labels = True)
plt.show()
```

```python
# other representations of graph
# an adjacency matrix: a matrix where individual nodes listed as columns and rows

print(nx.adjacency_matrix(G))
```

```python
# create adjacency matrix using Numpy
import numpy as np
G = nx.from_numpy_array(np.array([[0, 1, 0], [1, 1, 1], [0, 0, 0]]))
nx.draw_spring(G, with_labels = True)
plt.show()
```

```python
# determine the position of nodes, create a graph with the edge list we took above
edge_list = [(1, 2), (2, 3), (3, 4), (3, 5), (4, 6), (6, 7)]
G = nx.from_edgelist(edge_list)
```

```python
# draw the graph with a spring layout
nx.draw_spring(G, with_labels = True)
plt.show()
```

```python
# to arrange nodes in a circular way
nx.draw_circular(G, with_labels = True)
plt.show()
```

```python
# to do concentric circles
nx.draw_shell(G, with_labels = True)
plt.show()
```

```python
# with a spectral 2D layout
nx.draw_spectral(G, with_labels = True)
plt.show()
```

```python
# completely random way
nx.draw_random(G, with_labels = True)
plt.show()
```

```python
# as a planer graph: edges do not intersect
nx.draw_planar(G, with_labels = True)
plt.show()
```
